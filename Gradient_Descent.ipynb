```python
import matplotlib.pyplot as plt

# Function to minimize
def f(w):
    return (w - 3)**2
```

```python
# Derivative of the function (gradient)
def grad_f(w):
    return 2 * (w - 3)
```

```python
# Gradient Descent Algorithm with gradient tracking
def gradient_descent(starting_point, learning_rate=0.1, n_iterations=20):
    w = starting_point
    w_history = [w]
    grad_history = [grad_f(w)]

    for i in range(n_iterations):
        gradient = grad_f(w)
        w = w - learning_rate * gradient
        w_history.append(w)
        grad_history.append(grad_f(w))
        print(f"Step {i+1}: w = {w:.4f}, f(w) = {f(w):.4f}, gradient = {gradient:.4f}")

    return w_history, grad_history

# Run the algorithm
w_vals, grad_vals = gradient_descent(starting_point=0.0, learning_rate=0.1, n_iterations=20)
```

```python
iterations = list(range(len(w_vals)))
plt.plot(iterations, w_vals, marker='o')
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.title("Convergence of Gradient Descent")
plt.grid(True)
plt.show()
```

```python
# Plot 1: Value of x over iterations
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(len(w_vals)), w_vals, marker='o')
plt.title("Convergence of w")
plt.xlabel("Iteration")
plt.ylabel("w value")
plt.grid(True)

# Plot 2: Gradient values over iterations
plt.subplot(1, 2, 2)
plt.plot(range(len(grad_vals)), grad_vals, marker='s', color='red')
plt.axhline(0, color='black', linestyle='--')
plt.title("Gradient at Each Iteration")
plt.xlabel("Iteration")
plt.ylabel("Gradient Value")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Determine gradient sign
colors = []
for g in grad_vals:
    if abs(g) < 1e-3:
        colors.append('black')  # near zero
    elif g > 0:
        colors.append('red')    # positive
    else:
        colors.append('blue')   # negative

# Plotting gradient values with sign colors
plt.figure(figsize=(10, 5))
plt.axhline(0, color='gray', linestyle='--', label="Zero Gradient")
plt.scatter(range(len(grad_vals)), grad_vals, c=colors, s=100, edgecolors='k')
plt.plot(range(len(grad_vals)), grad_vals, linestyle='--', alpha=0.5)

# Labels and legend
plt.title("Gradient Sign Change Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Gradient Value")
plt.grid(True)
plt.legend(["Zero Gradient", "Gradient Points"])
plt.show()
```

```python
# Gradient descent with large step size
def gradient_descent(starting_point, learning_rate=1.0, n_iterations=15):
    w = starting_point
    w_vals = [w]
    grad_vals = [grad_f(w)]

    for i in range(n_iterations):
        gradient = grad_f(w)
        w = w - learning_rate * gradient
        w_vals.append(w)
        grad_vals.append(grad_f(w))
        print(f"Step {i+1}: w = {w:.4f}, f(w) = {f(w):.4f}, gradient = {gradient:.4f}")

    return w_vals, grad_vals

# Start on left side (x=0), large step size
w_vals, grad_vals = gradient_descent(starting_point=0.0, learning_rate=1.0, n_iterations=15)

# Color by gradient sign
colors = []
for g in grad_vals:
    if abs(g) < 1e-3:
        colors.append('black')  # near zero
    elif g > 0:
        colors.append('red')    # positive
    else:
        colors.append('blue')   # negative

# Plot gradient values
plt.figure(figsize=(10, 5))
plt.axhline(0, color='gray', linestyle='--', label="Zero Gradient")
plt.scatter(range(len(grad_vals)), grad_vals, c=colors, s=100, edgecolors='k')
plt.plot(range(len(grad_vals)), grad_vals, linestyle='--', alpha=0.5)

# Labels
plt.title("Gradient Sign Changes Over Iterations (Overshooting Minimum)")
plt.xlabel("Iteration")
plt.ylabel("Gradient Value")
plt.grid(True)
plt.legend(["Zero Gradient"])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Gradient descent
def gradient_descent(start_w, lr=1.0, iterations=15):
    w_vals = [start_w]
    for _ in range(iterations):
        grad = grad_f(w_vals[-1])
        new_w = w_vals[-1] - lr * grad
        w_vals.append(new_w)
    return w_vals

# Gradient descent trajectory
w_history = gradient_descent(start_w=0.0, lr=0.8, iterations=10)

# Function curve for plotting
x_plot = np.linspace(-3, 9, 400)
y_plot = f(x_plot)

# Set up the figure
fig, ax = plt.subplots(figsize=(8, 5))
point, = ax.plot([], [], 'o', markersize=10)
arrow = ax.annotate('', xy=(0, 0), xytext=(0, 0),
                    arrowprops=dict(facecolor='gray', shrink=0.05, width=2))
text = ax.text(0.02, 0.95, '', transform=ax.transAxes)

def init():
    ax.plot(x_plot, y_plot, 'k-', label='f(w) = (w - 3)^2')
    ax.axhline(0, color='gray', linestyle='--')
    ax.axvline(3, color='gray', linestyle=':')
    ax.set_xlim(-2, 8)
    ax.set_ylim(-1, 20)
    ax.set_title("Gradient Descent Animation")
    ax.set_xlabel("w")
    ax.set_ylabel("f(w)")
    ax.grid(True)
    return point, arrow, text

def update(i):
    w = w_history[i]
    y = f(w)
    grad = grad_f(w)

    # Arrow direction and color
    dw = 0.5 * np.sign(grad)
    dy = grad * dw
    color = 'red' if grad > 0 else 'blue' if grad < 0 else 'green'

    # Update point
    point.set_data([w], [y])
    point.set_color(color)

    # Update gradient arrow
    arrow.wy = (w + dw, y + dy)
    arrow.set_position((w, y))
    arrow.arrowprops['facecolor'] = color

    # Gradient value text
    text.set_text(f"Iteration {i}\nw = {w:.2f}\nGradient = {grad:.2f}")

    return point, arrow, text

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=len(w_history), init_func=init,
                              blit=True, repeat=False)

#  To save as .gif or .mp4
ani.save('gradient_sign_change.gif', writer='pillow', fps=1)
#ani.save('gradient_sign_change.mp4', writer='ffmpeg', fps=1)

plt.show()
```
